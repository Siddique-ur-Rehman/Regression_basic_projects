**Predicting salaries on single feature of experience:**

importing important libraries:


```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt
```

Loading file as Data frame:


```python
df=pd.read_csv('/content/Salary_dataset.csv')
df.head(4)
```

Data preprocessing required steps according to data need:

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

since data has no duplicate and null values to handle it we directly goes to selecting features and splitting data:

```python
X=df[['YearsExperience']] #feature
y=df['Salary'] #label
#splitting data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.15)
```

**Model Training**

```python
model1=LinearRegression()
model1.fit(X_train,y_train)
```

Evaluation metrices:

```python
from sklearn.metrics import mean_squared_error,r2_score
y_predict=model1.predict(X_test)
print(f'Mean Squared error: {mean_squared_error(y_test,y_predict)}')
print(f'R2 score: {r2_score(y_test,y_predict)}')
```

**Visual Representation:**

```python
plt.figure(figsize=(8,5))
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.scatter(X_test, y_test, color='green', label='Test Data')
plt.plot(X_train, model1.predict(X_train), color='red', linewidth=3, label='Regression Line')
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience (Feature)')
plt.ylabel('Salary (label / target)')
plt.grid(True)
plt.legend()
plt.show()
```
